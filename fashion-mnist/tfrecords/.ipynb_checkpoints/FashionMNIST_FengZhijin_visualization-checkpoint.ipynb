```python
# coding: utf-8
# author: Fengzhijin
# time: 2017.11.16
# ==================================
'''
实现了将FashionMNIST数据集tfrecords文件部分可视化检验文件是否正确
1.CATE - FashionMNIST标签字典
'''


import tensorflow as tf
import matplotlib.pyplot as plt

# fashionmnist数据集标签所代表的物品
CATE = {0: 'T-shirt/top', 1: 'Trouser', 2: 'Pullover', 3: 'Dress', 4: 'Coat',
        5: 'Sandal', 6: 'Shirt', 7: 'Sneaker', 8: 'Bag', 9: 'Ankle boot'}

if __name__ == "__main__":
    # tf.train.string_input_producer函数自动生成tfrecords文件名序列
    filename_queue = tf.train.string_input_producer(["../data/tfrecords/train.tfrecords"])
    # 创建一个reader来读取数据
    reader = tf.TFRecordReader()
    # 从tfrecords文件中读取一个样例
    _, serialized_example = reader.read(filename_queue)
    # 解析读入的一个样例
    features = tf.parse_single_example(
        serialized_example,
        features={
            'label': tf.FixedLenFeature([], tf.int64),
            'image_raw': tf.FixedLenFeature([], tf.string)
        })
    # tf.decode_raw函数将图像数据字符串解析成图相对应的像素数组
    image = tf.decode_raw(features['image_raw'], tf.uint8)
    label = features['label']

    with tf.Session() as sess:
        init_op = tf.global_variables_initializer()
        sess.run(init_op)
        # 启动多线程处理输入数据
        coord = tf.train.Coordinator()
        threads = tf.train.start_queue_runners(coord=coord)
        num = 9
        # 将读出的数据进行图像绘制
        for j in range(num):
            img, lab = sess.run([image, label])
            plt.subplot(3, num/3, j+1)
            lab = int(lab)
            plt.title(CATE[lab])
            plt.imshow(img.reshape([28, 28]), cmap='gray')
        plt.show()
        # 关闭线程
        coord.request_stop()
        coord.join(threads)
```
